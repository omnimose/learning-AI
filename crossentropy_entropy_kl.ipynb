Gibbs' Inequality
=================
Gibbs' inequality is a fundamental result in information theory and statistical mechanics that states that the Kullback-Leibler divergence (or relative entropy) between two probability distributions is always non-negative. It can be expressed mathematically as follows:
$$
D_{KL}(P || Q) = \sum_{x} P(x) \log \frac{P(x)}{Q(x)} \geq 0$$

where:  
- \( P \) and \( Q \) are two probability distributions over the same set of events.
- \( D_{KL}(P || Q) \) is the Kullback-Leibler divergence from \( P \) to \( Q \).


Sure! Let's **formally prove Gibbs' inequality**:

---

## üîí **Goal: Prove Gibbs' Inequality**

Let $p(x)$ and $q(x)$ be probability distributions over a discrete set $X$, with:

* $p(x) > 0$ for all $x \in X$,
* $q(x) > 0$ for all $x \in X$.

Then:

$$
\sum_{x \in X} p(x) \log \frac{p(x)}{q(x)} \geq 0
$$

with equality **if and only if** $p(x) = q(x)$ for all $x \in X$.

This is the KL divergence:

$$
D_{\text{KL}}(p \| q) \geq 0
$$

---

## üîß Proof using **log inequality**:

We use the elementary inequality:

> For all $a > 0$:
>
> $$
> \log a \leq a - 1 \quad \text{with equality if and only if } a = 1
> $$

Let‚Äôs apply this to $a = \frac{q(x)}{p(x)}$, or equivalently $\frac{p(x)}{q(x)}$.

---

### Step-by-step:

Let‚Äôs start with the KL divergence:

$$
D_{\text{KL}}(p \| q) = \sum_{x \in X} p(x) \log \frac{p(x)}{q(x)}
$$

Now apply the inequality $\log z \leq z - 1$ with $z = \frac{p(x)}{q(x)}$:

$$
\log \frac{p(x)}{q(x)} \geq 1 - \frac{q(x)}{p(x)}
$$

Multiply both sides by $p(x) > 0$:

$$
p(x) \log \frac{p(x)}{q(x)} \geq p(x) - q(x)
$$

Now sum over all $x$:

$$
\sum_x p(x) \log \frac{p(x)}{q(x)} \geq \sum_x \left(p(x) - q(x)\right)
= 1 - 1 = 0
$$

‚úîÔ∏è So:

$$
D_{\text{KL}}(p \| q) \geq 0
$$

---

### üü∞ When is equality achieved?

Equality in the log inequality occurs only when $\frac{p(x)}{q(x)} = 1 \Rightarrow p(x) = q(x)$

So:

$$
D_{\text{KL}}(p \| q) = 0 \quad \text{if and only if} \quad p(x) = q(x) \text{ for all } x
$$

---

## ‚úÖ Final Statement

$$
\boxed{
\sum_x p(x) \log \frac{p(x)}{q(x)} \geq 0
}
\quad \text{(with equality iff } p = q \text{)}
$$

```python
# -lnx >= 1 - x for x > 0

import numpy as np
import matplotlib.pyplot as plt

# Define the x values from just above 0 to 10
x = np.linspace(0.01, 10, 1000)  # More points for a smoother curve

# Define the functions
f1 = -np.log(x)        # f(x) = -ln(x)
f2 = 1 - x             # f(x) = 1 - x

# Plot the functions
plt.figure(figsize=(8, 5))
plt.plot(x, f1, label=r'$f(x) = -\ln x$', color='blue')
plt.plot(x, f2, label=r'$f(x) = 1 - x$', color='red', linestyle='--')

# Mark the intersection point at x = 1
plt.axvline(1, color='gray', linestyle=':', linewidth=1)
plt.scatter([1], [0], color='black')
plt.text(1.1, 0.2, 'x = 1', fontsize=10)

# Add labels and legend
plt.title('Plot of $f(x) = -\\ln x$ and $f(x) = 1 - x$ over $x \in (0, 10)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.ylim(-3, 3)
plt.xlim(0, 10)
plt.show()

```

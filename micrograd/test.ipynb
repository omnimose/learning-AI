```python
from graphviz import Digraph
from IPython.display import display

def draw_dag():
    # Create a new directed graph
    dag = Digraph(format='png', name="DAG Example")
    dag.attr(rankdir='LR', size='8,5')  # Set left-to-right layout
    
    # Add nodes
    dag.node('A', 'Start')
    dag.node('B', 'Step 1')
    dag.node('C', 'Step 2')
    dag.node('D', 'Step 3')
    dag.node('E', 'Finish')
    
    # Add edges
    dag.edges([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E')])
    dag.edge('B', 'D')  # Additional path
    
    # Render the graph to a file and display
    # dag.render('dag_example', view=True)
    return dag
```

```python
dag = draw_dag()
display(dag)
```

```python
float32_info = torch.finfo(torch.float32)
print("float32: ", float32_info)
float64_info = torch.finfo(torch.float64)   
print("float64: ", float64_info)
float16_info = torch.finfo(torch.float16)
print("float16: ", float16_info)
bfloat16_info = torch.finfo(torch.bfloat16)
print("bfloat16:", bfloat16_info)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# float32 constants
eps = 1.1920929e-07
resolution = 1e-06

# Create x values near 0 and 1
x_near_zero = np.linspace(-1e-5, 1e-5, 100)
x_near_one = np.linspace(0.9999999, 1.0000001, 100)

# Add resolution and eps to those x values
y_zero_plus_res = x_near_zero + resolution
y_one_plus_eps = x_near_one + eps

# Plot
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Left plot: resolution near 0
axs[0].plot(x_near_zero, y_zero_plus_res, label='x + resolution', color='blue')
axs[0].plot(x_near_zero, x_near_zero, '--', label='x (identity)', color='gray')
axs[0].set_title('Effect of resolution near 0 (float32)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].legend()
axs[0].grid(True)

# Right plot: eps near 1
axs[1].plot(x_near_one, y_one_plus_eps, label='x + eps', color='green')
axs[1].plot(x_near_one, x_near_one, '--', label='x (identity)', color='gray')
axs[1].set_title('Effect of eps near 1 (float32)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

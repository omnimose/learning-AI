```python
import torch
import math
```

```python
# Define the positional encoding function
def positional_encoding(length, d_model):
    pe = torch.zeros(length, d_model)
    freq = torch.zeros(length, d_model)
    for pos in range(length):
        for i in range(0, d_model, 2):
            pe[pos, i] = math.sin(pos / (10 ** ((2 * i) / d_model)))
            freq[pos, i] = pos / (10 ** ((2 * i) / d_model))
            if i + 1 < d_model:
                pe[pos, i + 1] = math.cos(pos / (10 ** ((2 * i) / d_model)))
                freq[pos, i + 1] = pos / (10 ** ((2 * i) / d_model))
    return pe, freq

# Define the parameters
length = 16  # The number of positions (sequence length)
d_model = 32  # The dimension of the embedding vector

# Calculate positional encodings
positional_encodings, freq = positional_encoding(length, d_model)

# Print the positional encodings
print("Positional Encodings:\n", positional_encodings)
```

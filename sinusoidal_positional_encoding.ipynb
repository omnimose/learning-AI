```python
import torch
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the positional encoding function
def positional_encoding(context_len, d_model):
    base = 100
    pe = torch.zeros(context_len, d_model)
    thetas = torch.zeros(context_len, d_model)
    for pos in range(context_len):
        for i in range(0, d_model):
            k = i // 2
            # print(pos, i, k)
            theta = pos / (base ** (2 * k / d_model))
            thetas[pos, i] = theta
            if i % 2 == 0:
                pe[pos, i] = math.sin(theta)
            else:
                pe[pos, i] = math.cos(theta)
    return pe, thetas

# Define the parameters
context_len = 8    # The number of positions (sequence length)
d_model = 32      # The dimension of the embedding vector

# Calculate positional encodings
positional_encodings, thetas = positional_encoding(context_len, d_model)

# Print the positional encodings
# print("Positional Encodings:\n", positional_encodings)
```

```python
# Convert positional encodings to a NumPy array for visualization
positional_encodings_np = positional_encodings.numpy()

# Plot the heatmap using Seaborn
plt.figure(figsize=(32, 8))
sns.heatmap(positional_encodings_np, cmap="viridis", annot=False)
# sns.heatmap(positional_encodings_np, cmap="viridis", annot=True, fmt=".2f", cbar=True)
plt.title("Positional Encodings Heatmap")
plt.xlabel("Embedding Dimension")
plt.ylabel("Sequence Position")
plt.show()
```

```python
# Convert positional encodings to a NumPy array for visualization
thetas_np = thetas.numpy()

# Plot the heatmap using Seaborn
plt.figure(figsize=(32, 8))
# sns.heatmap(freq_np, cmap="viridis", annot=False)
sns.heatmap(thetas_np, cmap="viridis", annot=True, fmt=".4f", cbar=True)
plt.title("Freq Heatmap")
plt.xlabel("Embedding Dimension")
plt.ylabel("Sequence Position")
plt.show()
```
